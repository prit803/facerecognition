```python
import face_recognition as fr
import cv2
import numpy as np
import os

def ima():
    cam = cv2.VideoCapture(0)
    result,image=cam.read()
    name=input("enter name:")
    if result:
        cv2.imshow("name:",image)
        cv2.imwrite(f"C:/Users/91953/Documents/Desktop/country/{name}.png",image)
    
    else:
        print("truy")

        
    # video_capture.release()
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return name
def check():
    video_capture=cv2.VideoCapture(0)
    u1=[]

    path="C:/Users/91953/Documents/Desktop/country/"


    known_flags=[]
    known_flags_encodings=[]

    images=os.listdir(path)

    for _ in images:
        images=fr.load_image_file(path + _)
        image_path=path + _
        encoding=fr.face_encodings(images)[0]
    
        known_flags_encodings.append(encoding)
        known_flags.append(os.path.splitext(os.path.basename(image_path))[0])
    print(known_flags)

    while True:
        _,frame=video_capture.read()
        small_frame=cv2.resize(frame,(0,0),fx=0.25,fy=0.25)
        rgb_small_frame=cv2.cvtColor(small_frame,cv2.COLOR_BGR2RGB)

        #recognize faces
    
        face_locations=fr.face_locations(rgb_small_frame)
        face_encodings=fr.face_encodings(rgb_small_frame)
    
        for face_encoding in face_encodings:
            matches=fr.compare_faces(known_flags_encodings,face_encoding)
            face_distance=fr.face_distance(known_flags_encodings,face_encoding)
            best_match_index=np.argmin(face_distance)
        
            if(matches[best_match_index]):
                name = known_flags[best_match_index]
            
            if name in known_flags:
                font=cv2.FONT_HERSHEY_SIMPLEX
                bottomLeftCornerOfText=(10,100)
                fontScale=1.5
                fontColor=(255,0,0)
                thickness=3
                lineType=2
                cv2.putText(frame,name +"present",bottomLeftCornerOfText,font,fontScale,fontColor,thickness,lineType)
                
            
        cv2.imshow("attendence:",frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return name

while(True):
    cust=[]
    user=input("Hello How Can I Help You\n\n  1.Create Account\n 2.Cash Credite\n 3.Cash Withdraw\n 4.Check Bal\n\n")
    if(user=='1'):
            a1=ima()
            print(f"Dear Customer, Your Account Created By The Name of {a1}")
            

            bal=0
            print(cust)
    elif(user=='2'):
        a2=check()
        print(a2)
        if a2==a1:
            amount=int(input("enter amount :"))
            bal=bal+amount;
        else:
            print("records not found")
    elif(user=='3'):
        a2=check()
        print(a2)
        if a2==a1:
            amount=int(input("enter amount :"))
            if(bal>amount):
                bal=bal-amount;
            else:
                print("Fund is Less")
        else:
            print("records not found")
    elif(user=='4'):
        a2=check()
        print(a2)
        if a2==a1:
            print(f"Your Total Amunt is {bal} Rupees")
        else:
            print("records not found")
    else:
        print("Thank you !")

```
